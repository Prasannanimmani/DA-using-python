```python
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv("/content/insurance.csv")

data['sex'] = data['sex'].replace({'female': 0, 'male': 1})
data['smoker'] = data['smoker'].replace({'no': 0, 'yes': 1})
data['region'] = data['region'].replace({'southwest': 0, 'northeast': 1,'southeast':2,'northwest':3})


print(data)

age_vector = data['age']
charge_vector = data['expenses']

region_factor = pd.Categorical(data['region'])

insurance_list = [
    age_vector,
    region_factor,
    data['sex'],
    data['expenses']
]

insurance_matrix = np.array(data[['age', 'bmi', 'expenses']])

insurance_df = data.copy()

print("Vector (age):")
print(age_vector.head())
print("\nFactor (region):")
print(region_factor.categories)
print("\nList (combination of vectors):")
print(insurance_list[:2])
print("\nMatrix (numeric columns):")
print(insurance_matrix[:5])
print("\nData Frame:")
print(insurance_df.head())

```

```python
import matplotlib.pyplot as plt

# Generate a scatter plot
plt.figure(figsize=(12, 6))

# Scatter plot: Age vs Expenses
age=data[data['age']<20]
print(age)

plt.scatter(data['age'], data['expenses'], c='blue', marker='.', alpha=0.7)

# Customizations
plt.title('Age vs Expenses', fontsize=16)  # Adding a title
plt.xlabel('Age', fontsize=12)            # X-axis label
plt.ylabel('Expenses', fontsize=12)       # Y-axis label
plt.grid(alpha=0.3)                       # Adding a light grid

# Add text annotations for specific points
for i in range(5):  # Annotate the first 5 points
    plt.text(data['age'][i], data['expenses'][i],
             f"({data['age'][i]}, {data['expenses'][i]})",
             fontsize=9, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
